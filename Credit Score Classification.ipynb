# Convert target column (Credit_Score) to numerical values (0-Poor, 1-Standard, 2-Good)

```python
df['Credit_Score'] = df['Credit_Score'].apply(lambda x: 0 if x == 'Poor' else 1 if x == 'Standard' else 2)
```

```python
df.info()
```

# Convert String column values to Numerical, using LabelEncoding

```python
# import LabelEncoder class
from sklearn.preprocessing import LabelEncoder

# create an instance of LabelEncoder
le = LabelEncoder()

# transform our string columns
df['Occupation'] = le.fit_transform(df['Occupation'])
df['Type_of_Loan'] = le.fit_transform(df['Type_of_Loan'])
```

```python
df
```

# Train-Split our Dataset

```python
# import train_split module
from sklearn.model_selection import train_test_split

# X will be used for training the ML model (not including the target column)
X = df.drop('Credit_Score', axis=1)

# target column
y = df['Credit_Score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Scale the data (only the X values (X_train, X_test)

```python
# import StandardScaler class
from sklearn.preprocessing import StandardScaler

# create an instance of StandardScaler
scaler = StandardScaler()

# scale the data
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Create ML model (KNN first)

```python
# import KNN model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# create an instance of KNN model
knn = KNeighborsClassifier(n_neighbors=5)  # number of K

# fit the model
knn.fit(X_train, y_train)

# model accuracy
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'KNN accuracy: {accuracy}')
```

# Metrics for KNN

```python
# import metrics
from sklearn.metrics import f1_score, recall_score, precision_score

knn_recall = recall_score(y_test, y_pred, average='macro')
knn_precision = precision_score(y_test, y_pred, average='macro')
knn_f1 = f1_score(y_test, y_pred, average='macro')

print(f'KNN Recall: {knn_recall}')
print(f'KNN Precision: {knn_precision}')
print(f'KNN F1: {knn_f1}')
```

# SVM model

```python
# import SVM class
from sklearn.svm import SVC

# create an instance of SVM
svc = SVC()

# fit the model
svc.fit(X_train, y_train)

# predictions
y_pred = svc.predict(X_test)

# accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'SVM accuracy: {accuracy}')
```

# Metrics for SVM

```python
# import metrics
from sklearn.metrics import f1_score, recall_score, precision_score

svm_recall = recall_score(y_test, y_pred, average='macro')
svm_precision = precision_score(y_test, y_pred, average='macro')
svm_f1 = f1_score(y_test, y_pred, average='macro')

print(f'SVM Recall: {svm_recall}')
print(f'SVM Precision: {svm_precision}')
print(f'SVM F1: {svm_f1}')
```

# Random Forest Model

```python
# Import random forest model
from sklearn.ensemble import RandomForestClassifier

# create an instance of random forest classifier
rf = RandomForestClassifier()

# fit the model
rf.fit(X_train, y_train)

# predictions
y_pred = rf.predict(X_test)

# accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Random Forest accuracy: {accuracy}')
```

# Metrics for RF

```python
# import metrics
from sklearn.metrics import f1_score, recall_score, precision_score

rf_recall = recall_score(y_test, y_pred, average='macro')
rf_precision = precision_score(y_test, y_pred, average='macro')
rf_f1 = f1_score(y_test, y_pred, average='macro')

print(f'RF Recall: {rf_recall}')
print(f'RF Precision: {rf_precision}')
print(f'RF F1: {rf_f1}')
```

# Naive Bayes Model

```python
# import NB model
from sklearn.naive_bayes import GaussianNB

# create an instance of NB
nb = GaussianNB()

# fit the model
nb.fit(X_train, y_train)

# predictions
y_pred = nb.predict(X_test)

# accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Naive Bayes accuracy: {accuracy}')
```
